```python
import numpy as np
import plotly.graph_objects as go

elevations = pow(np.linspace(0,1,100),2)*90

redScale = pow(elevations/90,0.1) * 255
greenScale = pow(elevations/90,0.4) * 255
blueScale = pow(elevations/90,2) * 255

intensityScale = (redScale + greenScale + blueScale)/3

colors = []
for i,el in enumerate(elevations):
    colors.append("rgb(" + '{:f}'.format(redScale[i]) + "," + '{:f}'.format(greenScale[i]) + "," + '{:f}'.format(blueScale[i])+")")

fig = go.Figure()
fig.add_trace(go.Scatter(x=elevations, y=redScale, line = dict(color='red')))
fig.add_trace(go.Scatter(x=elevations, y=greenScale, line = dict(color='green')))
fig.add_trace(go.Scatter(x=elevations, y=blueScale, line = dict(color='blue')))

fig.show()

fig = go.Figure()
fig.add_trace(go.Scatter(x=elevations, y=intensityScale, mode='markers', marker={'color':colors , 'colorscale': 'Rainbow', 'size': 10}))
fig.show()
```
